# Cleaning

Some image files seemed to be corrupted and were giving errors during trainng so I had to remove them.


```
train_dir, test_dir = "notMNIST_large", "notMNIST_small"
```

```
def delete_unreadable_images(path):
  for p, subdir, files in os.walk(path):
    for file in files:
      file_path = os.path.join(p, file)
      try:
        a = Image.open(file_path)
        arr = np.asarray(a)
      except IOError:
        os.remove(file_path)
        print(f"{file_path} deleted")
```

```
delete_unreadable_images(test_dir)
```

```
delete_unreadable_images(train_dir)
```

# Model Building

## Creating the tabular data set

```
label_dict = {letter: id for letter, id in zip(list("ABCDEFGHIJ"), range(10))}
label_dict
```

```
def create_table(path, which, label_dict):
  labels = []
  arrays = []
  for p, subdir, files in os.walk(path):
    for file in files:
      file_path = os.path.join(p, file)
      try:
        arr = np.asarray(Image.open(file_path)).reshape(-1)
        arrays.append(arr)
        labels.append(label_dict[p.split("/")[-1]])

      except Exception as e:
        print(str(e))
  return arrays, labels
```

```
test_array, test_labels = create_table(test_dir, "test", label_dict)
train_array, train_labels = create_table(train_dir, "train", label_dict)

print(np.array(test_array).shape, np.array(train_array).shape)
print(len(test_labels), len(train_labels))

train_d = np.array(train_array)
test_d = np.array(test_array)

train_l = tf.keras.utils.to_categorical(np.array(train_labels), 10)
test_l = tf.keras.utils.to_categorical(np.array(test_labels), 10)

print(train_l.shape, test_l.shape)
```

#### Shuffling the data

```
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]
```

```
train_d_shuffled, train_l_shuffled =  unison_shuffled_copies(train_d, train_l)
```

## Logistic Regression

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(10, input_dim=784, activation='softmax')) 

model.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics='acc',
              optimizer='adam')

history = model.fit(
  train_d_shuffled,
  train_l_shuffled,
  batch_size=32,
  epochs=10,
  validation_split = 0.2,
  verbose=1)
```

## CNN Model

Reshaping Input data to be Images to be fed to the CNN

```
train_d_img = train_d.reshape((-1, 28, 28))
test_d_img = test_d.reshape((-1, 28, 28))

train_d_img_shuffled, train_l_shuffled =  unison_shuffled_copies(train_d_img, train_l)
```

```
train_d_img[0].shape
```

```
shape = (28, 28, 1)
img_in = tf.keras.layers.Input(shape=shape, name='input')
x = tf.keras.layers.Dropout(0.2, name='input_dropout')(img_in)

# conv block 1
x = tf.keras.layers.Conv2D(16, (3,3), activation='relu')(x)
x = tf.keras.layers.Conv2D(32, (3,3), activation='relu')(x)
x = tf.keras.layers.Conv2D(64, (3,3), activation='relu')(x)
x = tf.keras.layers.MaxPooling2D((2,2))(x)
block_1 = tf.keras.layers.Dropout(0.5)(x)

# conv block 2
x = tf.keras.layers.Conv2D(64, (3,3), activation='relu')(block_1)
x = tf.keras.layers.Conv2D(128, (3,3), activation='relu')(x)
x = tf.keras.layers.Conv2D(128, (3,3), activation='relu')(x)
x = tf.keras.layers.MaxPooling2D((2,2))(x)
block_2 = tf.keras.layers.Dropout(0.5)(x)

# GAP Layer
X = tf.keras.layers.GlobalAveragePooling2D()(block_2)
output = tf.keras.layers.Dense(10, activation='softmax')(X)


# compile model
model = tf.keras.models.Model(img_in, output)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```
history = model.fit(
  train_d_img_shuffled,
  train_l_shuffled,
  batch_size=32,
  epochs=10,
  validation_split = 0.2,
  verbose=1)
```
